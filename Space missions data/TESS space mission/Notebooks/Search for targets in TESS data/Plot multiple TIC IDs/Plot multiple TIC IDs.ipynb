# The below code is to produce normalized light curves

```python
#Plots multiple TIC IDs from a CSV file and save all flattened lightcurves that their flux vary in a given range 
#in a single pdf file. Also, the saved TICIDs are stored in a new CSV file
#All the skipped TICIDs are saved in a new CSV file and plotted in a different pdf file.
#If an 502 Server Error: Bad Gateway occurs from MAST, the script will catch it, print a message and skip it

import lightkurve as lk
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np
import time
import requests

df = pd.read_csv("Sector60 - one.csv")
tic_ids = df["TICID"].tolist()

valid_tic_ids = []
skipped_tic_ids = []
start_time = time.time()

with PdfPages('Sector60 - TIC IDs.pdf') as valid_pdf:
    with PdfPages('Sector60 - Skipped TIC IDs.pdf') as skipped_pdf:
        for i, tic_id in enumerate(tic_ids):
            print("Processing TIC ID: ",tic_id, " Progress: ", i+1, " of ", len(tic_ids))
            try:
                lc = lk.search_lightcurve("TIC " + str(tic_id), author="SPOC",sector=60, exptime=120).download()
            except requests.exceptions.HTTPError as err:
                if err.response.status_code == 502:
                    print("Error downloading TIC ID: ", tic_id, " Skipping to next target.")
                    continue
                else:
                    raise
            normalized_lc = lc.normalize()
            if (normalized_lc.flux > 0.95).all() & (normalized_lc.flux < 1.05).all():
                valid_tic_ids.append(tic_id)
                normalized_lc.plot(color = 'black', lw = 0, marker = '.',label='TIC {} Normalized'.format(tic_id))
                plt.legend(fontsize=8)
                valid_pdf.savefig()
                plt.clf()
            else:
                skipped_tic_ids.append(tic_id)
                lc.plot(color = 'black', lw = 0, marker = '.',label='TIC {}'.format(tic_id))
                plt.legend(fontsize=8)
                skipped_pdf.savefig()
                plt.clf()
                print(f"Skipping TIC ID: {tic_id} because some flux values are not between 0.95 and 1.05.")
        # create a new csv file with valid TIC IDs
        valid_tic_df = pd.DataFrame(valid_tic_ids, columns=["TICID"])
        valid_tic_df.to_csv("tic_ids.csv", index=False)
        # create a new csv file with skipped TIC IDs
        skipped_tic_df = pd.DataFrame(skipped_tic_ids, columns=["TICID"])
        skipped_tic_df.to_csv("skipped_tic_ids.csv", index=False)

end_time = time.time()
elapsed_time = end_time - start_time
print("Elapsed Time: {:.2f} seconds".format(elapsed_time))
```
