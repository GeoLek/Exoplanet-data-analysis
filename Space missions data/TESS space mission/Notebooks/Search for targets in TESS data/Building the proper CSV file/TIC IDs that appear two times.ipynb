This script searches through all the CSV files in the current working directory and compares the values in each column between the files. The user is prompted to select one file as the "main" file to search in. Then, the script checks how many times each unique value in each column of the main file appears in the other files. If a value appears twice, it is added to a new DataFrame, which is then written to a new CSV file named "Sectorxx - two.csv", where xx is the number of the sector you search. At the end, the script prints out the elapsed time of the script.

```python
#Check all TIC IDs in a CSV file of your choice
#compare them with the TICIDs from all the other CSV files in the same folder
#and save the TICIDs that appear only two times in other files, in a new CSV file


import os
import pandas as pd
import time

start_time = time.time()


def search_and_compare_values():
    files = [f for f in os.listdir() if f.endswith('.csv')]
    for i, file in enumerate(files):
        print(f'{i+1}. {file}')
    main_file_index = int(input('Please select the file you want to search in by entering its corresponding number: ')) - 1
    main_file_name = files[main_file_index]
    df_main = pd.read_csv(main_file_name)
    new_file_data = []
    for column in df_main.columns:
        for value in df_main[column].unique():
            count = 0
            for file_name in files:
                if file_name != main_file_name:
                    df = pd.read_csv(file_name)
                    if value in df[column].values:
                        count += 1
            if count == 2:
                new_file_data.append({'TICID': value})
    new_file_df = pd.DataFrame(new_file_data)
    new_file_df.to_csv('Sector59 - two.csv', index=False)

search_and_compare_values()



end_time = time.time()
elapsed_time = end_time - start_time
print("Time elapsed: ", elapsed_time)
```
